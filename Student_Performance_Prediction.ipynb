```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("student_data.csv")

# Convert Result to numbers
data['Result'] = data['Result'].map({'Fail': 0, 'Pass': 1})

# Features and target
X = data[['Study_hours', 'Attendence', 'Previous_Marks']]
y = data['Result']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction
study_hours = float(input("Enter Study Hours: "))
attendance = float(input("Enter Attendance: "))
marks = float(input("Enter Previous Marks: "))

new_data = pd.DataFrame([[study_hours, attendance, marks]],
                        columns=['Study_hours', 'Attendence', 'Previous_Marks'])

prediction = model.predict(new_data)

if prediction[0] == 1:
    print("Student will PASS")
else:
    print("Student will FAIL")
# Accuracy
y_pred = model.predict(X_test)
print("Model Accuracy:", accuracy_score(y_test, y_pred))
```
